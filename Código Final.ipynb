### **Estadisticas de mayor peso por posición**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
import pickle

df= pd.read_csv(output_path_normalized)

# Función para seleccionar características relevantes
def seleccionar_caracteristicas(posicion, df, features_por_posicion):
    relevant_features = features_por_posicion.get(posicion, []) + features_por_posicion['Any']
    return df[relevant_features]

# Función para entrenar un modelo RandomForest y obtener las importancias de las características
def obtener_importancias(posicion, df, features_por_posicion):
    subset = df[df['Posicion'] == posicion]
    if subset.shape[0] == 0:
        return f"No hay suficientes datos para la posición {posicion}."
    
    X = seleccionar_caracteristicas(posicion, subset, features_por_posicion)
    y = subset['Calificacion']
    
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X, y)
    
    importancias = model.feature_importances_
    feature_names = X.columns
    
    return pd.Series(importancias, index=feature_names).sort_values(ascending=False)

# Visualización de las 10 características más importantes para cada posición
def visualizar_importancias(importancias, posicion):
    top_10_importancias = importancias.head(10)
    plt.figure(figsize=(10, 6))
    top_10_importancias.plot(kind='bar', color='skyblue')
    plt.title(f'Top 10 Estadísticas más importantes para {posicion}')
    plt.xlabel('Estadística')
    plt.ylabel('Importancia')
    plt.show()

# Diccionario para almacenar las 10 características más importantes por posición
top_10_features_por_posicion = {}

# Obtener y visualizar importancias para cada posición
posiciones = ['Goalkeeper', 'Defender', 'Midfielder', 'Forward']
for posicion in posiciones:
    importancias = obtener_importancias(posicion, df, features_por_posicion)
    if isinstance(importancias, pd.Series):
        top_10_features_por_posicion[posicion] = importancias.head(10).index.tolist()
        visualizar_importancias(importancias, posicion)
    else:
        print(importancias)

# Guardar las 10 características más importantes por posición en un archivo
with open('top_10_features_por_posicion.pkl', 'wb') as file:
    pickle.dump(top_10_features_por_posicion, file)

# Cargar las características más importantes de un archivo
with open('top_10_features_por_posicion.pkl', 'rb') as file:
    top_10_features_por_posicion = pickle.load(file)

# Mostrar las características más importantes por posición
print("Top 10 características más importantes por posición:")
for posicion, features in top_10_features_por_posicion.items():
    print(f"{posicion}: {features}")
```
